```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.linear_model import SGDClassifier 
from sklearn.model_selection import train_test_split, GridSearchCV, validation_curve, learning_curve, cross_val_score, RandomizedSearchCV
from sklearn.svm import SVC, LinearSVC
from sklearn.metrics import confusion_matrix , precision_recall_curve,plot_precision_recall_curve
from sklearn.pipeline import make_pipeline 
from sklearn.tree import  DecisionTreeClassifier
from sklearn.preprocessing import RobustScaler, StandardScaler, PolynomialFeatures,label_binarize
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_selection import SelectKBest, f_classif #f_classif pour le test d'Anova
from sklearn.multiclass import OneVsRestClassifier
```

```python
rawdata= pd.read_excel('newBagdata.xls')
```

```python
#classer les donnees du debit en (0= forage sec , 1= 0-1; 2=1-3 m3/h , 3= 3-6 m3/h , 4=6-10 m3/h ,5= +10 m3/h )
df1=rawdata.copy()
df1.loc[df1['debitm']==0,'debitm']=0
df1.loc[(df1['debitm']>0) & (df1['debitm']<=3), "debitm"]=1
# df1.loc[(df1['debitm']>0) & (df1['debitm']<=1), "debitm"]=1
# df1.loc[(df1['debitm']>1) & (df1['debitm']<=3), "debitm"]=2 
df1.loc[(df1['debitm']>3) &(df1['debitm']<=6), "debitm"]=2
df1.loc[(df1['debitm']>6) & (df1['debitm']<=10), "debitm"]=3 
df1.loc[df1['debitm']>10, "debitm"]=4

df1.set_index('nfor', inplace=True)
df1.head()
```

```python
# Faut encoder les valeurs categories dans notre data set 
 
df =df1.copy()
df['forme'] = df['forme'].astype('category').cat.codes
df['type'] = df['type'].astype('category').cat.codes
df['geol'] = df['geol'].astype('category').cat.codes
X=df.drop(['x_m', 'y_m','naEau','debitm'], axis=1)
y=df['debitm']
df.head()
```

```python
# construction du graphique de debit 
```

```python
# Diviser le dataset entre Trainset et Testset  (decocher cela pour le modeling )
X_train, X_test, y_train, y_test=train_test_split(X, y,test_size=0.25, random_state=7)
# df.head(50)
```

```python
# On va entrainer un modele de SVM pour voir sa performance sans processing 
model1=SVC(random_state=7)
model1.fit(X_train, y_train)
model1.score(X_test, y_test)
```

```python
#Ameloirer ce modele pour voir sa conduite a la recherche de meilleur parametre 
hyper_params={'C': [1, 10, 100, 1000], 
             'gamma':[1e-2, 1e-3, 1e-4]}
grid=GridSearchCV(model1, hyper_params, cv=4)
grid.fit(X_train, y_train)
grid.best_score_
grid.best_params_
```

```python
#sauvegarde le model a partir de Grid searchCV
model1=grid.best_estimator_
model1.score(X_test, y_test)
```

```python
#creation de pipeline 
preprocessor = make_pipeline(PolynomialFeatures(2, include_bias=False),RobustScaler(), SelectKBest(k=6))
```

```python
#modele
model_svm=make_pipeline(preprocessor, SVC(C=1, gamma=1e-4, random_state=7))
model_tree =make_pipeline(preprocessor, DecisionTreeClassifier(max_depth=100, random_state=7))
model_rand =make_pipeline(preprocessor, RandomForestClassifier(max_depth=100, random_state=7))
model_ada =make_pipeline(preprocessor, AdaBoostClassifier(random_state=7))
model_knn =make_pipeline(preprocessor, KNeighborsClassifier(n_neighbors = 7))
```

```python
#Entrainer chaque modele et voire le resultats 
model_svm.fit(X_train, y_train)
model_tree.fit(X_train, y_train)
model_rand.fit(X_train, y_train)
model_ada.fit(X_train, y_train)
model_knn.fit(X_train, y_train)
#Affichage des scores 
print(model_svm.score(X_test, y_test))
print(model_tree.score(X_test, y_test))
print(model_rand.score(X_test, y_test))
print(model_ada.score(X_test, y_test))
print(model_knn.score(X_test, y_test))
```

```python
# recherche de meilleurs parametres pour le modele avec Grid Search CV
hyper_params ={'svc__C': [1, 10, 100],
               'svc__gamma':[1e-1, 1e-2, 1e-3]}

grid_svm=GridSearchCV(model_svm, hyper_params, cv=4)

grid_svm.fit(X_train, y_train)
grid_svm.best_score_
grid_svm.best_params_
```

```python
#sauvegarde le model a partir de Grid searchCV
model=grid_svm.best_estimator_
model.score(X_test, y_test)
```

```python
#Booster le model 
model_final =AdaBoostClassifier( n_estimators=100,  random_state=7)
model_final.fit(X_train, y_train)
model_final.score(X_test, y_test)
```

```python
# Rechercher de meilleurs parametres sur ploynomial Features et SelectKbest 

hyperparams ={'pipeline__polynomialfeatures__degree': np.arange(1,20),
              'pipeline__selectkbest__k':np.arange(1,8)}

grid_svm2=RandomizedSearchCV(model_svm, hyperparams, cv=4)

grid_svm2.fit(X_train, y_train)
print(grid_svm2.best_params_)
model3=grid_svm.best_params_
```
