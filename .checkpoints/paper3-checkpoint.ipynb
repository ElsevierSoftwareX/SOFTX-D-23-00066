```python
data_cat=data_raw.copy()
data_cat['debitm'].map(lambda x:x+1)
#debitm=x+1
def cat_debitm(debitm):
    if debitm==0:
        return 'forage sec'
    elif debitm>0 and debitm<=1: 
        return '0-1 m3/h'
    elif debitm>1 and debitm<=3: 
        return '1-3 m3/h'
    elif debitm>3 and debitm<=6: 
        return '3-6 m3/h'
    elif debitm>6 and debitm<=10: 
        return '6-10 m3/h'
    else : 
        return '+10 m3/h'
data_cat['debitm']=data_cat['debitm'].map(cat_debitm)
data_cat.head()
```

```python
g = sns.catplot(x="forme", hue="debitm", col="type",
...                 data=data_cat, kind="count",
                    linewidth=5,
...                 height=4, aspect=.7);
plt.show()
plt.savefig('formTypedebit_1.png')
g = sns.catplot(x="type", hue="debitm", col="geol",
...                 data=data_cat, kind="count",
                    linewidth=5,
...                 height=4, aspect=.7);
plt.show()
plt.savefig('typegeoldebit_1.png')
```
